# Data Preprocessing

```python
# Drop rows where total is missing
df = df.dropna(subset=["total"])
df
```

```python
# Keep only the final row of each match (last ball), because it represents the final score
final_df = df.groupby("mid").tail(1).copy()
final_df
```

```python
#tail(1) after groupby("mid") means: “take the last record for each match ID”.
```

```python
# Create a new column 'winner' — same as batting_team if total is highest in the match
#We will label the winner as the team with the highest total.
match_totals = final_df.groupby("mid")["total"].idxmax()
match_totals
```

```python
# 1. Group by match and team to get their total score
team_totals = df.groupby(['mid', 'batting_team'])['total'].max().reset_index()
team_totals
```

```python
# 2. Find the team with max score per match — they are the winner
match_winners = team_totals.loc[team_totals.groupby('mid')['total'].idxmax()].reset_index(drop=True)
match_winners.rename(columns={'batting_team': 'winner'}, inplace=True)
```

```python
match_winners
```

```python
#This merges the correct winner column to each match's last ball record.

# 3. Take only the final record of each match (for model input) and add the winner
final_df = df.groupby("mid").tail(1).copy()
final_df = final_df.merge(match_winners[['mid', 'winner']], on='mid')
final_df
```

```python
features = final_df[["batting_team", "bowling_team", "venue", "total", "wickets", "overs", "runs_last_5", "wickets_last_5"]]
labels = final_df["batting_team"] == final_df["winner"]  # binary: True if team won
features
labels
```

# Encode the text data

```python
le = LabelEncoder()
for col in ["batting_team", "bowling_team", "venue"]:
    features[col] = le.fit_transform(features[col])
le
```

# Train the Model

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

# Evaluate

```python
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
